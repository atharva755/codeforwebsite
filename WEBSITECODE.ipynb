```python
import streamlit as st

# Sample product data with images
products = [
    {"name": "Smart TV", "category": "Electronics", "price": 450, "rating": 4.5, "image": "https://via.placeholder.com/150"},
    {"name": "Running Shoes", "category": "Footwear", "price": 70, "rating": 4.7, "image": "https://via.placeholder.com/150"},
    {"name": "Laptop", "category": "Electronics", "price": 800, "rating": 4.8, "image": "https://via.placeholder.com/150"},
]

# CSS for styling
st.markdown(
    """
    <style>
        .product-card {
            border-radius: 10px;
            background-color: #f9f9f9;
            padding: 20px;
            margin: 10px 0;
            box-shadow: 0px 4px 8px rgba(0, 0, 0, 0.1);
        }
        .product-card img {
            width: 100%;
            border-radius: 10px;
        }
        .product-details {
            text-align: center;
        }
        .add-cart-button {
            background-color: #ff4b4b;
            color: white;
            border: none;
            padding: 10px;
            border-radius: 5px;
            cursor: pointer;
            font-size: 16px;
            margin-top: 10px;
        }
        .add-cart-button:hover {
            background-color: #ff2b2b;
        }
    </style>
    """,
    unsafe_allow_html=True
)

# Sidebar Filters
st.sidebar.title("Filters")
category = st.sidebar.selectbox("Category", ["All", "Electronics", "Footwear", "Clothing"])
price_range = st.sidebar.slider("Price Range", 0, 1000, (0, 500))
search_query = st.sidebar.text_input("Search Products", "")

# Main Content
st.title("üõí Fancy E-Commerce Website")
st.subheader("Discover and shop your favorite products!")

# Filtered Product List
st.write("### Products")
filtered_products = [
    product for product in products 
    if (category == "All" or product["category"] == category) 
    and (price_range[0] <= product["price"] <= price_range[1])
    and (search_query.lower() in product["name"].lower() if search_query else True)
]

# Display Products as Cards
for product in filtered_products:
    # Product Card
    st.markdown(
        f"""
        <div class="product-card">
            <img src="{product['image']}" alt="{product['name']}">
            <div class="product-details">
                <h3>{product['name']}</h3>
                <p>Category: {product['category']}</p>
                <p>Price: ${product['price']}</p>
                <p>Rating: {"‚≠ê" * int(product['rating'])} ({product['rating']})</p>
                <button class="add-cart-button" onClick="alert('Added {product['name']} to cart!')">Add to Cart</button>
            </div>
        </div>
        """,
        unsafe_allow_html=True
    )

# Shopping Cart Summary (optional, can be implemented using session states)
st.sidebar.subheader("Shopping Cart")
# Placeholder for future cart implementation
st.sidebar.text("No items in cart.")
```
